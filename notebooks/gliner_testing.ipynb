```python
import sys
sys.path.insert(0, '/content/gliner-sports-ner')
from gliner import GLiNER

model = GLiNER.from_pretrained("checkpoints/gliner-sports/best", load_tokenizer=True)

test_text = "Donovan Mitchell scored 32 points as the Cavaliers defeated the Raptors 126-113."
entities = model.predict_entities(test_text, [
    "PLAYER", "TEAM", "STAT", "GAME_EVENT"
], threshold=0.5)

for e in entities:
    print(f"[{e['label']}] {e['text']}")
```

```python
# Test base model load in isolation
from gliner import GLiNER
import time

print("Loading base model...")
start = time.time()
model = GLiNER.from_pretrained("urchade/gliner_medium-v2.1")
print(f"Loaded in {time.time()-start:.1f}s")

# Quick test
entities = model.predict_entities(
    "LeBron James scored 30 points for the Lakers",
    ["PLAYER", "TEAM", "STAT"],
    threshold=0.5
)
print(entities)
```

```python
from gliner import GLiNER
import time

print("Loading fine-tuned model...")
start = time.time()
ft_model = GLiNER.from_pretrained(
    "checkpoints/gliner-sports/best",
    load_tokenizer=True
)
print(f"Loaded in {time.time()-start:.1f}s")

entities = ft_model.predict_entities(
    "LeBron James scored 30 points for the Lakers",
    ["PLAYER", "TEAM", "STAT"],
    threshold=0.5
)
print(entities)
```
